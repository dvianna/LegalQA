````python
import pandas as pd
````

````python
# load portuguese legal QA data
dataset = pd.read_csv("LegalQA_dataset.csv", sep = '\t')
````

````python
dataset.shape
````

````python
dataset.head()
````

````python
def buildprompt(data):
    prompt = {}
    prompt['prompt'] = "Given the question delimited by triple backticks ```{" + data['question'] + "}```, what is the answer?"
    print(data['answer'])
    prompt['completion'] = "{" + data['answer'] + "}"
    return prompt
````

````python
dataset.dropna()
dataset.shape
````

````python
dataset['prompt'] = dataset.apply(buildprompt, axis=1)
````
